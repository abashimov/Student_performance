# **Modelling**

### **Task1: Predictive Modeling**

The project proceeds with the construction of a predictive model using the `RandomForestRegressor` from the `sklearn.ensemble` module. The model is built for Task 1, which involves predicting the `G1.Math` variable.

Hyperparameter tuning is performed using `GridSearchCV` from the `sklearn.model_selection` module. The hyperparameters tuned are `n_estimators` and `max_depth`, with the possible values being `[100, 200, 300]` and `[None, 5, 10]` respectively. `GridSearchCV` performs an exhaustive search over the specified parameter values for the estimator, and uses cross-validation with `cv=5` to avoid overfitting.

The model is then trained on the numeric training data `X_train_task1_numeric` and the corresponding labels `y_train_task1`. The best parameters found by `GridSearchCV` are printed, providing insights into the optimal configuration of the `RandomForestRegressor` for this specific task.

The model's performance is evaluated using the Root Mean Squared Error (RMSE), a common metric for regression tasks. The RMSE is computed on the test data `X_test_task1_numeric` and the corresponding true labels `y_test_task1`. The predicted labels are obtained using the `predict` method of the trained model. The RMSE value is then printed, providing a quantitative measure of the model's performance on unseen data. This completes the model building and evaluation process for Task 1.

````python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV

# Build a predictive model using RandomForestRegressor
# Use GridSearchCV for hyperparameter tuning
parameters_task1 = {'n_estimators': [100, 200, 300], 'max_depth': [None, 5, 10]}
````

The code imports essential libraries (Optional) for building a predictive model, including RandomForestRegressor for regression tasks, mean_squared_error for performance evaluation, and GridSearchCV for hyperparameter tuning.
A dictionary parameters_task1 is created, specifying different values for the hyperparameters 'n_estimators' and 'max_depth' to be explored during the hyperparameter tuning process.

````python
# Use cross-validation in GridSearchCV
model_task1 = GridSearchCV(RandomForestRegressor(), parameters_task1, cv=5)
````

GridSearchCV is initialized with the RandomForestRegressor model and the hyperparameter grid (parameters_task1). Cross-validation with 5 folds (cv=5) is specified to assess model performance across different subsets of the training data.

````python
model_task1.fit(X_train_task1_numeric, y_train_task1)
````

The GridSearchCV model is trained on the numeric features (X_train_task1_numeric) and corresponding target variable (y_train_task1).

````python
# Print best parameters
print(f'Task 1 - Best Parameters: {model_task1.best_params_}')
````

The code prints the best hyperparameters determined by the GridSearchCV process.

````python
# Evaluate the model using RMSE
y_pred_task1 = model_task1.predict(X_test_task1_numeric)
rmse_task1 = mean_squared_error(y_test_task1, y_pred_task1, squared=False)
print(f'Task 1 - RMSE: {rmse_task1}')
````

The trained model is used to predict the target variable on the test set (X_test_task1_numeric), and the root mean squared error (RMSE) is calculated as a measure of the model's predictive performance. The RMSE is then printed for assessment.

### **Task2: Classification Modeling**

Task 2 of the project involves building a classification model for the variable `G1.Math`, which is binned into three categories: 'low', 'medium', and 'high'. The binning is performed using the `qcut` function from the pandas library, which ensures that each bin has approximately the same number of observations.

The dataset is then preprocessed similarly to Task 1, with unnecessary variables being dropped, missing values being handled, and categorical variables being converted into numerical form. The data is then split into training and testing sets, and the numerical features are scaled.

A `RandomForestClassifier` from the `sklearn.ensemble` module is used to build the classification model. Hyperparameter tuning is performed using `GridSearchCV`, with the parameters `n_estimators` and `max_depth` being tuned. The model is trained on the numeric training data and the corresponding labels.

The best parameters found by `GridSearchCV` are printed, providing insights into the optimal configuration of the `RandomForestClassifier` for this task. The model's performance is evaluated using accuracy, which is computed on the test data and the corresponding true labels.

Additional evaluation metrics are also computed for this classification task. A confusion matrix is computed, which provides a summary of the correct and incorrect predictions made by the classifier. A classification report is also generated, which includes precision, recall, and F1-score for each class. These metrics provide a more comprehensive view of the model's performance, taking into account both the positive and negative classes. This completes the model building and evaluation process for Task 2. The model is now ready to be used for making predictions on new data.

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Task 2: Classification Model for Binned G1.Math
# Bin the target variable G1.Math into 3 categories
df['G1.Math.Binned'] = pd.qcut(df['G1.Math'], q=3, labels=['low', 'medium', 'high'])
````

Import essential libraries for building a classification model, including RandomForestClassifier for classification tasks and metrics like accuracy_score, precision_score, recall_score, f1_score, and confusion_matrix for model evaluation.
The target variable 'G1.Math' is binned into three categories ('low', 'medium', 'high') using pandas qcut function.

````python
# Filter out the unnecessary variables for Task 2
df_task2 = df.drop(['address', 'sex', 'Medu', 'G2.Math', 'G3.Math', 'G1.Math'], axis=1)
````

Unnecessary variables are dropped from the dataset for Task 2 to focus on relevant features.

````python
# Check for missing values and handle them appropriately for Task 2
df_task2 = df_task2.dropna()
````

Missing values in the filtered dataset are dropped to ensure data quality.

````python
# Convert categorical variables into numerical form for Task 2
df_task2[categorical_cols] = df_task2[categorical_cols].apply(lambda col: le.fit_transform(col))
````

Categorical variables are encoded into numerical form using LabelEncoder for Task 2.

````python
# Split the data into training and testing sets for Task 2
X_task2 = df_task2.drop('G1.Math.Binned', axis=1)
y_task2 = df_task2['G1.Math.Binned']
X_train_task2, X_test_task2, y_train_task2, y_test_task2 = train_test_split(X_task2, y_task2, test_size=0.2, random_state=42)
````

he dataset is split into training and testing sets for Task 2.

````python
# Scale the features for Task 2
X_train_task2_numeric = scaler_task1.fit_transform(X_train_task2.select_dtypes(include=['number']))
X_test_task2_numeric = scaler_task1.transform(X_test_task2.select_dtypes(include=['number']))
````

The numeric features for Task 2 are scaled using the same scaler used for Task 1.

````python
# Build a classification model using RandomForestClassifier
# Use GridSearchCV for hyperparameter tuning
parameters_task2 = {'n_estimators': [100, 200, 300], 'max_depth': [None, 5, 10]}
# Use cross-validation in GridSearchCV
model_task2 = GridSearchCV(RandomForestClassifier(), parameters_task2, cv=5)
model_task2.fit(X_train_task2_numeric, y_train_task2)
````

GridSearchCV is employed to find the best hyperparameters for the RandomForestClassifier model using cross-validation.

````python
# Print best parameters
print(f'Task 2 - Best Parameters: {model_task2.best_params_}')
````

The best hyperparameters determined by the GridSearchCV process are printed.

````python
# Evaluate the model using accuracy
y_pred_task2 = model_task2.predict(X_test_task2_numeric)
accuracy_task2 = accuracy_score(y_test_task2, y_pred_task2)
print(f'Task 2 - Accuracy: {accuracy_task2}')
````

The classification model is evaluated using accuracy, and the result is printed.

````python
# Additional Evaluation Metrics for Classification Task (Task 2)

# Confusion Matrix
conf_matrix = confusion_matrix(y_test_task2, y_pred_task2)
print(f'\nConfusion Matrix:\n{conf_matrix}')
````

The confusion matrix is printed to provide detailed information about the model's performance on the test set.

````python
# Use classification report for Task 2
from sklearn.metrics import classification_report
print(classification_report(y_test_task2, y_pred_task2))
````

The classification report is printed, which includes precision, recall, and F1-score for each class in the target variable. This provides a comprehensive evaluation of the model's performance.
